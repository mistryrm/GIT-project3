```python
import pandas as pd
import re

import requests
import json
from pprint import pprint
import time
```

```python
# review cleanned data
data_df = pd.read_json("../data/met_data.json").transpose()

# read map for country .json
path = open("../data/original_data/map_countries.json")
map_json = json.load( path )
map_df = pd.json_normalize(map_json['ref_country_codes'])
map_df.head()
```

```python
# find all country used in met data
country_df = data_df['country'].value_counts().to_frame( "count").reset_index( drop = False )
country_df.columns = ["country","count"]
country_df = pd.merge( country_df, map_df , how="left" , on="country" )
country_df
```

```python
# save as csv for manual adjust
map_df.to_csv( "../data/original_data/map_list.csv" , index = False , na_rep="" )
country_df.to_csv( "../data/original_data/country_list.csv" , index = False , na_rep="" )
```

```python
# conver organized map to json file
map_df = pd.read_csv( "../data/original_data/map_point.csv" )

map_json = {}

for i in range(len(map_df)):
    temp_geo = { "latitude": map_df.loc[ i , 'latitude' ] , "longitude": map_df.loc[ i , 'longitude' ] }
    map_json[ map_df.loc[ i , 'geoCode' ] ] = temp_geo
    
with open("../data/map_point.json", "w") as outfile:
    json.dump(map_json, outfile)
```
